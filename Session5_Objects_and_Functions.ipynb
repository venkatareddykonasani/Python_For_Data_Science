## map() and filter()

1. `map()` -> `map(func, iterable)`

  Creates an map object(iterator) that computes a given `func` with arguments from iterable.
2. `filter()` -> `filter(func, iterable)`

  Returns iterator with items of the iterable for which func is `True`

### map()

```
list1 = [1,5,6,7,8]

tmp_fn = lambda x : x**2
```

```
# tmp_fn(list1)
list2 = [tmp_fn(i) for i in list1]
list2
```

```
map(tmp_fn, list1)
```

```
map(tmp_fn, list1)
list(map(tmp_fn, list1))
```

### filter()

```
list1 = [1,2,3,4,5]
fil_fun = lambda x : x%2 == 0
```

```
filter(fil_fun, list1)
list(filter(fil_fun, list1))
```

# OOP: Object Oriented Programming
Super basic to get by.

## What are Objects?
objects are 'data abstractions':
1. Data               : **Attributes**
2. Play with the data : **Methods**


### Object analogy:
**Object**: think of a bike

* **Class** : The chasis of the bike/ structure

    1. **Attributes**: what represents this bike: handles, colour, headlight

    2. **Methods**: what can you do with it/interact: ride, horn, on/off light

```
class Employee(object):

    # Constructor
    def __init__(self, first, last, salary):
        self.first = first  # Create an instance variable
        self.last = last  # Create an instance variable
        self.salary = salary  # Create an instance variable
        self.email = f'{first}.{last}@company.com'
        
    # method
    def full_name(self):
        # return f'{first} {last}'
        return f'{self.first} {self.last}'
```

```
emp1 = Employee('stan', 'marsh', 1000)  # Construct an instance of the Greeter class
emp2 = Employee('eric', 'cartman', 2500)
```

```
emp1.first
```

```
print(emp1.full_name())
print(emp2.full_name())
```

#### Inheritance

This was not covered in class, but if you want to try this snippet and understand by yourself what's happening here, please do.

```
class Developer(Employee):
    pass
    
dev1 = Developer('stan', 'marsh', 1000)
print(dev1.full_name())
```

```
class Developer(Employee):
    def __init__(self, first, last, salary, plang):
        super().__init__(first, last, salary)
        self.plang = plang

dev1 = Developer('stan', 'marsh', 1000, ['java'])
```

```
dev1.full_name()
```
