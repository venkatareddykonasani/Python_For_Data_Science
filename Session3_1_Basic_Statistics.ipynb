#Variance and Standard Deviation

```
YouTubeVideo('5iFUYLAFm9c', width=500, height=300)
# https://youtu.be/5iFUYLAFm9c
```

```
Online_Retail=pd.read_csv("https://raw.githubusercontent.com/venkatareddykonasani/Datasets/master/Online%20Retail%20Sales%20Data/Online%20Retail.csv",encoding = "ISO-8859-1")
```

```
Online_Retail.info()
```

```
#var and sd UnitPrice
var_UnitPrice=Online_Retail['UnitPrice'].var()
print(var_UnitPrice)

std_UnitPrice=Online_Retail['UnitPrice'].std()
print(std_UnitPrice)
```

```
var_qty=Online_Retail['Quantity'].var()
print(var_qty)

std_qty=Online_Retail['Quantity'].std()
print(std_qty)
```
